This exercise uses the [Seinfeld Chronicles dataset](https://www.kaggle.com/thec03u5/seinfeld-chronicles). The objective is to create a dictionary of the most common words (that are not 'stop' words) used by each character during the series. First, I load in the data.

```python
def load_scripts(path_to_scripts):
    file_handle = open(path_to_scripts, "r")
    x = list(file_handle)
    file_handle.close
    return x
seinfeld = load_scripts("data/scripts.csv")
len(seinfeld), seinfeld[:10]
```

```python
import csv
def load_scripts_csv(path_to_scripts):
    with open(path_to_scripts) as S:
        reader = csv.reader(S)
        data= [tuple(row) for row in reader]
    return data

seinfeld = load_scripts_csv("data/scripts.csv")
len(seinfeld), [(x[1], x[2]) for x in seinfeld[-19:]]
```

Next, I collect the data from the `Character` column and load it into a list called `speakers`.

```python
def get_speakers(seinfeld):
    speakers = [col[1] for col in seinfeld[1:]]
    return speakers
speakers = get_speakers(seinfeld)
characters = set(speakers) #the unique values ONLY
len(characters), list(characters)[:25]
```

Next I count the dialogue entries for the four major characters (`"JERRY"`, `"GEORGE"`, `"ELAINE"`, and `"KRAMER"`), using  a dictionary. For this assignment, I was not allowed to use the `Counter`, and instead had to do it from scratch.

```python
def count_four_main(speakers):
    counts = {
        "JERRY" : 0,
        "GEORGE" : 0,
        "ELAINE" : 0,
        "KRAMER" : 0
    }
    for name in speakers:
        if name == 'JERRY' or name =='GEORGE' or name =='ELAINE' or name=='KRAMER':
            counts[name] += 1
        else:
            counts[name] = 1
    return counts
counts = count_four_main(speakers)
counts
```

Now I count the number of words spoken by each of the main characters using a dictionary.

```python
def count_words_spoken(seinfeld):
    words_spoken = {
        "JERRY" : 0,
        "GEORGE" : 0,
        "ELAINE" : 0,
        "KRAMER" : 0
    }
    for i in seinfeld:
        if i[1] not in words_spoken:
            words_spoken[i[1]] = len(i[2].split())
        else:
            words_spoken[i[1]] += len(i[2].split())
    return words_spoken
words_spoken = count_words_spoken(seinfeld)
words_spoken
```

```python
def count_words(seinfeld):
    word_counts = {
        "JERRY" : {},
        "GEORGE" : {},
        "ELAINE" : {},
        "KRAMER" : {}
    }
    for row in seinfeld:
        if row[1] in word_counts:
            temp = row[2].split(" ")
            for word in temp:
                if word in word_counts[row[1]]:
                    word_counts[row[1]][word] += 1
                else:
                    word_counts[row[1]][word] = 1                       
    return word_counts
```

```python
# who says yada, what's, and babka more often?
word_counts = count_words(seinfeld)
jerry_total = sum(word_counts["JERRY"].values())
elaine_total = sum(word_counts["ELAINE"].values())
[(word, word_counts["JERRY"].get(word, 0)/jerry_total, word_counts["ELAINE"].get(word, 0)/elaine_total)
 for word in ["yada", "yada", "what's", "babka"]]

#Elaine says "yada" more frequently
#Jerry says "babka" more frequently
```

Finally, I want to create a dictionary of the most common words for each of the main characters that are _not_ in the list of stop words. This will be sorted from high to low.

```python
stop_words = [word.strip() for word in open("data/stop-words.txt", "r").readlines()]

def get_top_N(word_counts, stop_words, N = 10):
    most_common_words = {
        "JERRY" : [],
        "GEORGE" : [],
        "ELAINE" : [],
        "KRAMER" : []
    }
    for character in word_counts:
        for word in word_counts[character]:
            if word.lower() not in stop_words:
                most_common_words[character].append((word.lower(), word_counts[character][word]))
   
        most_common_words[character].sort(key=lambda x:x[1], reverse = True)
        del most_common_words[character][N:]
    
    return most_common_words

most_common_words = get_top_N(word_counts, stop_words)
most_common_words
```
